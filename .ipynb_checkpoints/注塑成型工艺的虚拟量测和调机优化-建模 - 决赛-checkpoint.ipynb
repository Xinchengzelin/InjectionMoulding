# 模型构建

## 评估函数

```python
fig,ax=plt.subplots(2,1,figsize=(16,0.1))
def my_loss_1(st, sp, issize1=True):#st:真实尺寸，sp:预测尺寸
    st=np.array(st)[:,0]
    sp=np.array(sp)
    if issize1:
        LL = 299.85
        UL = 300.15
    else:
        LL = 199.925
        UL = 200.075
    w,b,e=np.ones(sp.shape[0]),np.zeros(sp.shape[0]),np.exp(abs(st - sp)/0.012) - 1
    for i in range(len(st)):
        if st[i] >= LL and st[i] <= UL:
            if sp[i] < LL or sp[i] > UL:
                w[i] = 10
        if st[i] <LL:
            b[i] = abs(st[i] - LL)
            if sp[i] >= LL:
                w[i] = 10
        else:
            b[i] = abs(st[i] - UL)
            if sp[i] <= UL:
                w[i] = 10
    a=100*b+1#样本异常系数
    score = np.sum(w*e*a)/len(sp)
    return score
def my_loss_1_1value(st, sp, issize1=True):#st:真实尺寸，sp:预测尺寸
    if issize1:
        LL = 299.85
        UL = 300.15
    else:
        LL = 199.925
        UL = 200.075
    w,b,e=1,0,np.exp(abs(st - sp)/0.012) - 1
    if st >= LL and st <= UL:
        if sp < LL or sp > UL:
            w = 10
    if st <LL:
        b = abs(st - LL)
        if sp >= LL:
            w = 10
    else:
        b = abs(st - UL)
        if sp <= UL:
                w = 10
    a=100*b+1#样本异常系数
    score = np.sum(w*e*a)
    return np.log10(score)

def plot_model_loss(model,x_train=x_train1,y_train=y_train1,train=train1,ax=ax,is_ypred_2d=False):
#     plt.subplots(2,1,figsize=(16,8))
    y_true=y_train.iloc[:,0]
    if not is_ypred_2d:
        y_pred=model.predict(x_train)
    else:
        y_pred=model.predict(x_train)[:,0]
    data=pd.DataFrame({"Id":train.Id,"true":y_true,"pred":y_pred})
    data["log_score"]=data.apply(lambda x:my_loss_1_1value(x.true,x.pred,issize1=True),axis=1)#
    data.sort_values("Id",inplace=True)
    ax[0].scatter(x=data.Id,y=data.true,c='b',marker=".",s=5)
    ax[0].scatter(x=data.Id,y=data.pred,c='r',marker=".",s=5)
    print("最大score的Id:%d,对应的log分数:%.8e"%(data[data.log_score==max(data.log_score)]["Id"],max(data.log_score)))
    ax[1].scatter(x=data.Id,y=data.log_score,c='r',marker="+",s=10)
    plt.show()
    return data
    
```

## size1

```python
%%time

pipe_lgb1_pca=Pipeline(memory=None,
     steps=[('pca',
              PCA(copy=True, iterated_power='auto', n_components=25, random_state=42,
                svd_solver='auto', tol=0.0, whiten=False)),
             ('lgb1',
              lgb.LGBMRegressor(bagging_fraction=0.8, bagging_freq=2, boosting_type='gbdt',
                     class_weight=None, colsample_bytree=1.0, feature_fraction=0.8,
                     importance_type='split', learning_rate=0.12, max_depth=12,
                     min_child_samples=20, min_child_weight=0.001, min_split_gain=0.0,
                     n_estimators=100, n_jobs=-1, num_iterations=200, num_leaves=63,
                     objective='regression', random_state=41, reg_alpha=0.0,
                     reg_lambda=0.0, silent=True, subsample=1.0,
                     subsample_for_bin=200000, subsample_freq=0))])
myfunc_loss = make_scorer(my_loss_1, greater_is_better=False)
scores = cross_val_score(pipe_lgb1_pca, x_train1, y_train1, cv=10, scoring=myfunc_loss)
print("**",scores,scores.mean())
pipe_lgb1_pca.fit(x_train1,y_train1)
print ('Training MAE score: %.4f' % mean_absolute_error(y_train1, pipe_lgb1_pca.predict(x_train1)))
print ('Training my_loss score: %.8E' % my_loss_1(y_train1, pipe_lgb1_pca.predict(x_train1), issize1=True))
```

```python
fig,ax=plt.subplots(2,1,figsize=(16,10))
data = plot_model_loss(pipe_lgb1_pca,x_train=x_train1,ax=ax)
data.sort_values("log_score",ascending=False)[:10]
```

## size2

```python
%%time
model_lgb2=lgb.LGBMRegressor(boosting_type= 'gbdt',  # 设置提升类型
                                objective= 'regression',  # 目标函数
                                num_leaves=32,  # 叶子节点数
                                learning_rate=0.05,  # 学习速率
                                feature_fraction=0.9,  # 建树的特征选择比例
                                bagging_fraction= 0.8,  # 建树的样本采样比例
                                bagging_freq= 5,  # k 意味着每 k 次迭代执行bagging
                                verbose= -1)

myfunc_loss = make_scorer(my_loss_1, issize1=False,greater_is_better=False)
scores = model_selection.cross_val_score(model_lgb2, x_train2, y_train2, cv=10, scoring=myfunc_loss)
print("**",scores,scores.mean())
model_lgb2.fit(x_train2, y_train2);
print ('Training MAE score: %.4f' % mean_absolute_error(y_train2, model_lgb2.predict(x_train2)))
print ('Training my_loss score: %.8e' % -my_loss_1(y_train2, model_lgb2.predict(x_train2),issize1=False))
```

```python
fig,ax=plt.subplots(2,1,figsize=(16,10))
data = plot_model_loss(model_lgb2,x_train=x_train2,y_train=y_train2,train=train2,ax=ax)
data.sort_values("log_score",ascending=False)[:10]
```

```python
#计算预测结果
def cal_pred_new(model1,model2,model3,test1=x_test1,test2=x_test2,test3=x_test3):
    sub_file["size1"]=model1.predict(test1)
    sub_file["size2"]=model2.predict(test2)
    sub_file["size3"]=model3.predict(test3)
    res = pd.concat([sub_file.loc[:,["size1","size2","size3"]].mean(axis=0),sub_file.loc[:,["size1","size2","size3"]].median(axis=0)],axis=1)
    res=res.rename(columns={0:"均值",1:"中位数"})
    print("预测均值/中位数：\n",res)
    return sub_file
```

```python
sub_file=pd.read_csv("./taskA/sub_file.csv")
sub_file=cal_pred_new(pipe_lgb1_pca,model_lgb2,model_lgb3,test1=x_test1_taskA,test2=x_test2_taskA,test3=x_test3_taskA)
sub_file.to_csv("./sub_file_taskA.csv",index=False)
sub_file.head()
```
